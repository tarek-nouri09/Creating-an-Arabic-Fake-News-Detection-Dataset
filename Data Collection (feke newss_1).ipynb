```python
from bs4 import BeautifulSoup
import requests
import re 
```

```python
tokens=[]
for i in range (1,143) :
    url = f'https://verify-sy.com/list/16/%D8%A7%D8%AD%D8%AA%D9%8A%D8%A7%D9%84?page={i}'
    page =requests.get(url)
    soup = BeautifulSoup(page.text ,'html')
    soup.find_all('div', class_='content-list-desc')
    
    for div in soup.find_all('div', class_='content-list-desc'):
        paragraph = div.find('p').get_text().strip().replace("الادعاء", "").replace("\n", "")
        paragraph = re.search(r'^(.*?\.)', paragraph, re.MULTILINE | re.UNICODE)
        tokens.append(paragraph.group(1))
        
print(tokens)
```

```python
import pandas as pd
df = pd.DataFrame({'text': tokens})
df['category'] = 'كذب'
print(df)
```
