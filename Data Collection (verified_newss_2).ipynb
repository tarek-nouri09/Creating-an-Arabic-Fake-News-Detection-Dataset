```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import pandas as pd 
```

```python
verifiednewss = []

for i in range(1, 41):
    link = "https://www.bbc.com/arabic/topics/cv2xyrnr8dnt?page=" + str(i)
    req = Request(link, headers={'User-Agent': 'Mozilla/5.0'})
    html_page = urlopen(req).read()
    soup = BeautifulSoup(html_page, 'html.parser')
    mydivs = soup.find_all("div", {"class": "promo-text"})

    def contains_text(text):
        return text and "المدة" in text

    for i in mydivs:
        to_remove_div = i.find('span')
        tags_to_remove = soup.find_all(string=contains_text)
        for tag in tags_to_remove:
            tag.parent.extract()
        verifiednewss.append(i.find("a").text)
```

```python
for i in range(len(verifiednewss)):  
    verifiednewss[i] = verifiednewss[i].replace("فيديو,", "", 1)

verifiednewss = [[x, '1'] for x in verifiednewss]

for i in range(len(verifiednewss)):  
    print(verifiednewss[i])
```

```python
df = pd.DataFrame(verifiednewss, columns=['News', 'State'])
df.to_csv('verified_newss_956.csv', index=False)
```
